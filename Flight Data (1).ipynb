```python
# Lets create the empty lists for the required columns
airline_name = []
journey_date = []
journey_day= []
source = []
destination = []
departure_time = []
arrival_time = []
duration = []
total_stops = []

fare = []
```

```python
# Extracting  airline_name
try:
    airline = driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')      
    
    for i in airline:
        airline_name.append(i.text)
        
except NoSuchElementException:
    airline_name.append('-')
    

    
# Appending same date corresponding to all flights
for j in range(94):
    journey_date.append('2022-02-08')
    
    day= driver.find_element_by_xpath('//div[@class="day-li text-center cursor-pointer pr active font-primary-color"]//p[1]')
    journey_day.append(day.text)

# Extracting  Source of the flight
try:
    source1 = driver.find_elements_by_xpath('//p[@title="Mumbai"]')      
    
    for i in source1:
        source.append(i.text)
        
except NoSuchElementException:
    source.append('-')
        
# Extracting  destination of the flight
try:
    destination1 = driver.find_elements_by_xpath('//p[@title="Lucknow"]')      
    
    for i in destination1:
        destination.append(i.text)
        
except NoSuchElementException:
    destination.append('-')
    
    
# Extracting  departure_time of the flight
try:
    departure_time1 = driver.find_elements_by_xpath('//div[@class="i-b pr"]')      
    
    for i in departure_time1:
        departure_time.append(i.text)
        
except NoSuchElementException:
    departure_time.append('-')
    
    
# Extracting  arrival_time of the flight
try:
    arrival_time1 = driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')      
    
    for i in arrival_time1:
        arrival_time.append(i.text)
        
except NoSuchElementException:
    arrival_time.append('-')
    
# Extracting  duration of the flight
try:
    duration1 = driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')      
    
    for i in duration1:
        duration.append(i.text)
        
except NoSuchElementException:
    duration.append('-')
    
# Extracting  total_stops of the flight
try:
    total_stops1 = driver.find_elements_by_xpath('//span[@class="dotted-borderbtm" or @class="cursor-default"]')      
    
    for i in total_stops1:
        total_stops.append(i.text)
        
except NoSuchElementException:
    total_stops.append('-')
    
# Extracting  fare of the flight
try:
    fare1 = driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')      
    
    for i in fare1:
        fare.append(i.text)
        
except NoSuchElementException:
    fare.append('-')



        
```

```python
Flight_data=pd.DataFrame({})
Flight_data["airline_name"]= airline_name
Flight_data["journey_date"]= journey_date
Flight_data["journey_day"] = journey_day
Flight_data["source"]= source
Flight_data["destination"]= destination
Flight_data["departure_time"]= departure_time
Flight_data["arrival_time"]= arrival_time
Flight_data["duration"]= duration
Flight_data["total_stops"]= total_stops
Flight_data["fare"]= fare
Flight_data
```

```python
Flight_data.to_csv('Flight.csv')
```
